# MSS vs. MVM

```python
## Imports
import nilearn as nil
from nilearn.plotting import plot_glass_brain
from nilearn import plotting

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Define path to mask
mask_path = '../ROI_masks/MAX_ROIs_final_gm.nii.gz'
mask = mask_path.split('/')[-1]
```

```python
# Load text with ROI info as a dataframe
roi_df = pd.read_csv('/data/bswift-1/Pessoa_Lab/MAX/ROI_masks/README_MAX_ROIs_final_gm.txt',sep='\t',index_col = 'Index')
roi_df['name'] = roi_df[['Hemi','ROI']].apply(lambda x: ' '.join(x),axis=1)
rois = roi_df.name
roi_df.head()
```

## Click around to look at ROIs

```python
# Load the ROI mask
mask = nil.image.load_img(mask_path)

view = plotting.view_img(mask_path,vmax=rois.index.max(),symmetric_cmap = False,)
# In a Jupyter notebook, if ``view`` is the output of a cell, it will
# be displayed below the cell
view
```

```python
# Load MSS and MVM response
mss_threat = np.squeeze(nil.image.load_img('/data/bswift-1/Pessoa_Lab/MAX/dataset/positive_runs/Group/t0.nii.gz').get_data())
mss_safe = np.squeeze(nil.image.load_img('/data/bswift-1/Pessoa_Lab/MAX/dataset/positive_runs/Group/s0.nii.gz').get_data())

mvm_threat = np.squeeze(nil.image.load_img('/data/bswift-1/Pessoa_Lab/MAX/dataset/positive_runs/Group/threat_bucket.nii.gz').get_data())
mvm_safe = np.squeeze(nil.image.load_img('/data/bswift-1/Pessoa_Lab/MAX/dataset/positive_runs/Group/safe_bucket.nii.gz').get_data())
```

## MSS vs. MVM in 58 ROIs

```python
fig, axes = plt.subplots(nrows=10, ncols = 6, figsize=(20,25))
axes = axes.flatten()
markersize=6
fig.subplots_adjust(wspace=0.2,hspace=0.35)

for ii, roi in rois.items():
    axes[ii-1].plot(mss_threat[mask.get_data() == ii].mean(axis=0),c='red',alpha=1,marker='o',markersize=markersize,label='MSS, threat')
    axes[ii-1].plot(mvm_threat[mask.get_data() == ii].mean(axis=0)[:12],c='red',alpha=1,marker='*',markersize=markersize,label='MVM, threat')

    axes[ii-1].plot(mss_safe[mask.get_data() == ii].mean(axis=0),c='blue',alpha=1,marker='o',markersize=markersize,label='MSS, safe')
    axes[ii-1].plot(mvm_safe[mask.get_data() == ii].mean(axis=0)[:12],c='blue',alpha=1,marker='*',markersize=markersize,label='MVM, safe')
    
    axes[ii-1].set_title(roi)
fig.legend(axes[ii-1].get_legend_handles_labels()[1],loc='center right',fontsize=16)
```
