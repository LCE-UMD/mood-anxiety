```python
from src.data import *
from src.prepare import *

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("data/MAX_threat_vs_safe.txt")
```

```python
fig, axes = plt.subplots(nrows=10,ncols=6,figsize=(15,30),sharey=True)
fig.subplots_adjust(wspace=0.25,hspace=0.4)
axes = axes.flatten()

for ii, roi in enumerate(df.ROI.unique()):
    axes[ii].set_title(roi)
    sns.boxplot(x='cond',y='beta',data=df[df['ROI']==roi],ax=axes[ii])
```

```python
from collections import defaultdict

corr_dict = defaultdict(dict)
cols = []
for subj in df.Subj.unique():
    subj_df = df[df['Subj']==subj]
    for cond in 'safe threat'.split():
        cond_df = subj_df[subj_df['cond'] == cond]
        corr_df = cond_df.drop('cond',axis=1).set_index(['Subj','Trial','ROI']).unstack(-1).corr('spearman')
        corr_df.columns = corr_df.columns.droplevel()
        corr_df.index = corr_df.index.droplevel()
        cols.append(corr_df.columns)
        corr_dict[subj][cond] = corr_df
        
safe = np.dstack([corr_dict[subj]['safe'] for subj in corr_dict])
threat = np.dstack([corr_dict[subj]['threat'] for subj in corr_dict])
threat_v_safe = threat - safe

cols = np.vstack(cols)
print((cols == cols[0]).all())

mean_safe = pd.DataFrame(np.mean(safe,axis=2),columns=cols[0],index=cols[0])
mean_threat = pd.DataFrame(np.mean(threat,axis=2),columns=cols[0],index=cols[0])
mean_threat_v_safe = pd.DataFrame(np.mean(threat_v_safe,axis=2),columns=cols[0],index=cols[0])
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(mean_threat_v_safe,vmin=-1.0,center=0,vmax=1.0,cmap='coolwarm',annot=True,fmt='.1f',annot_kws={"size":8})
_=plt.title('Threat vs Safe',fontsize=16)
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(mean_threat,vmin=-1.0,center=0,vmax=1.0,cmap='coolwarm',annot=True,fmt='.1f',annot_kws={"size":8})
_=plt.title('Mean Threat',fontsize=16)
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(mean_safe,vmin=-1.0,center=0,vmax=1.0,cmap='coolwarm',annot=True,fmt='.1f',annot_kws={"size":8})
_=plt.title('Mean Safe',fontsize=16)
```
