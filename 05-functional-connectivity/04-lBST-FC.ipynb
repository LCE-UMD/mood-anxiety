# Functional Connectivity (FC)
## Left BST 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
EOI = []

for effect in "Intercept cond".split():
    brms = pd.read_csv("results/MAX_threat_v_safe_L_BST_FC_{}_brms.txt".format(effect))
    glm = pd.read_csv("results/MAX_threat_v_safe_L_BST_FC_{}_glm.txt".format(effect))
    comb = brms[['P+']].join(glm[['2-sided-p']])
    
    EOI.append(comb)
```

## Intercept: Mean FC

```python
def imshow(path):
    img = mpimg.imread(path)
    plt.figure(figsize = (10,15))
    plt.imshow(img,interpolation='nearest')
```

```python
imshow('results/MAX_threat_v_safe_L_BST_FC_Intercept.png')
print(EOI[0])
```
